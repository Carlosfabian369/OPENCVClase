```
import cv2
import matplotlib.pyplot as plt
```

```
imagen = cv2.imread("/content/Imagen.png")
plt.imshow(imagen)
```

```
# Leer la imagen
image = cv2.imread("Imagen.png")

# Convertir a escala de grises
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Aplicar diferentes tipos de umbralización
_, img1 = cv2.threshold(gray_image, 50, 255, cv2.THRESH_BINARY)
_, img2 = cv2.threshold(gray_image, 50, 255, cv2.THRESH_BINARY_INV)
_, img3 = cv2.threshold(gray_image, 50, 255, cv2.THRESH_TRUNC)
_, img4 = cv2.threshold(gray_image, 50, 255, cv2.THRESH_TOZERO)
_, img5 = cv2.threshold(gray_image, 50, 255, cv2.THRESH_TOZERO_INV)

# Mostrar las imágenes resultantes usando matplotlib
plt.figure(0)
plt.imshow(img1, cmap='gray')  # Imágenes binarizadas (0 - 255)
plt.title('Umbral Binario')
plt.axis('off')  # Desactiva los ejes

plt.figure(1)
plt.imshow(img2, cmap='gray')  # Imágenes binarizadas invertidas
plt.title('Umbral Binario Invertido')
plt.axis('off')

plt.figure(2)
plt.imshow(img3, cmap='gray')  # Imágenes truncadas
plt.title('Umbral Truncado')
plt.axis('off')

plt.figure(3)
plt.imshow(img4, cmap='gray')  # Umbralización a cero
plt.title('Umbral a Cero')
plt.axis('off')

plt.figure(4)
plt.imshow(img5, cmap='gray')  # Umbralización a cero invertida
plt.title('Umbral a Cero Invertido')
plt.axis('off')

# Mostrar todas las figuras
plt.show()
```
